Soal 2

```python
vect_1 = np.array([17, 22, 19])
vect_2 = np.array([10, 20, 11])
vect_3 = np.array([5, 12, 9])

print(vect_1)
print(vect_2)
print(vect_3)
```

```python
print('A + B = \n' + str(vect_1 + vect_2))
```

```python
print('B - C = \n' + str(vect_2 - vect_3))
```

```python
print('A dot C = \n' + str(vect_1 @ vect_3))
```

```python
print('A x C = \n' + str(vect_1 * vect_2))
```

```python
print('norm A= \n' + str(np.linalg.norm(vect_1)))
```

```python
cos_theta = vect_1 @ vect_2 / (np.linalg.norm(vect_1) * np.linalg.norm(vect_2))
cos_theta
```

```python
print('Sudut antara vektor A dan vektor B adalah = ' + str(np.rad2deg(cos_theta)))
```

Soal 3

```python
u = np.array([2, 5])
v = np.array([3, 1])
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_vector2d(vector2d, origin=[0, 0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
              head_width=0.2, head_length=0.3, length_includes_head=True,
              **options)
```

```python
plot_vector2d(v, color='r')
plot_vector2d(u, color='g')

plt.axis([0,10,0,10]) # rentang axis sumbu x dan y
plt.grid()
plt.show()
```

```python
def konversi(suhu):             #membuat fungsi konversi
    tocelc = suhu - 273.15      #convert dari kelvin ke celcius
    tokelv = suhu + 273.15      #convert dari celcius ke kelvin
    return tocelc, tokelv
```

```python
#Print hasil convert
print(f'Suhu dari kelvin ke celcius adalah : {konversi(1)[0]}')
print(f'Suhu dari celcius ke kelvin adalah : {konversi(1)[1]}')
```

```python
def convtofahr(suhu):       #Membuat fungsi convert ke fahrenheit
    
    tofahr = konversi(suhu, suhu)               #memanggil fungsi convert sebelumnya 
    tofahr = list(tofahr)                       #Mengubah tuple menjadi list agar dapat dihitung per index-nya
    tofahr[0] = (tofahr[0] * 1.8) + 32          #Convert dari suhu kelvin ke fahrenheit
    tofahr[1] = ((tofahr[1] - 273.15)*1.8) + 32 #Convert dari suhu celcius ke fahrenheit

    return  tofahr

```

```python
#Print hasil convert

print(f'Suhu fahrenheit dari kelvin adalah  {convtofahr(1)[0]}')
print(f'Suhu fahrenheit dari celcius adalah  {convtofahr(1)[1]}')
```

```python
def convfromfahr(fahr):                     #Membuat fungsi convert dari fahrenheit
    tokel = ((fahr - 32) * 5/9) + 273.15    #Convert suhu dari fahrenheit ke kelvin
    tocel = ((fahr - 32)* 5/9)              #Convert suhu dari fahrenheit ke celcius

    return tokel, tocel

 
```

```python
#Print hasil convert

print(f'Convert suhu dari fahrenheit ke kelvin adalah = {convfromfahr(1)[0]}')
print(f'Convert suhu dari fahrenheit ke celcius adalah = {convfromfahr(1)[1]}')
```
